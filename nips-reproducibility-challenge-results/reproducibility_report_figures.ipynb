# Figures for Mini Project 4

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import itertools
sns.set(font_scale = 1.2)
sns.set_style("whitegrid")
def flip(items, ncol):
    return itertools.chain(*[items[i::ncol] for i in range(ncol)])
```

Figure 3 (Test Accuracy for various masking criteria)

```python
data = pd.read_csv('crawled_data_figure_3.csv', index_col = [0])
```

```python
data.head()
```

```python
data['weights_rem'] = (np.power(0.8, data['iteration']) * 235200 + 300 + np.power(0.8, data['iteration']) * 30000 + 100 + np.power(0.9, data['iteration']) * 1000 +10)/266610.0
data['weights_rem'] = data['weights_rem'] * 100
data.head()
```

```python
def _plot_range_band(*args, central_data=None, ci=None, data=None, **kwargs):
    upper = data.max(axis=0)
    lower = data.min(axis=0)
    #import pdb; pdb.set_trace()
    ci = np.asarray((lower, upper))
    kwargs.update({"central_data": central_data, "ci": ci, "data": data})
    sns.timeseries._plot_ci_band(*args, **kwargs)

sns.timeseries._plot_range_band = _plot_range_band

ax = sns.tsplot(time='weights_rem',value="test_accuracy", condition="mask_criterion", unit="seed", data=data,
               err_style="range_band", n_boot=0, legend = False)
ax.set(xscale="log")
ax.invert_xaxis()
ax.set_ylim(0.95, 0.99)
ax.set(xlabel='Weights Remaining', ylabel='Test Accuracy')
#ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

ax.set_xlim(100,1.2)
ax.xaxis.set_major_formatter(mtick.PercentFormatter())
plt.locator_params(nbins=6, axis = 'y')
plt.xticks([100.0, 41.0, 16.8, 6.9, 2.8, 1.2])
_ = [ax.get_children().pop(x).remove() for x in [6,4, 3, 2, 1]]
```

```python
sns.timeseries._plot_range_band = _plot_range_band
ax1 = sns.tsplot(time='weights_rem',value="test_accuracy", condition="mask_criterion", unit="seed", data=data,
               err_style="range_band", n_boot=0)
ax1.set(xscale="log")
ax1.invert_xaxis()
ax1.set_ylim(0.975, 0.985)
ax1.set(xlabel='Weights Remaining', ylabel='Test Accuracy')
ax1.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

ax1.set_xlim(100,1.2)
ax1.xaxis.set_major_formatter(mtick.PercentFormatter())
plt.locator_params(nbins=6, axis = 'y')
plt.xticks([100.0, 41.0, 16.8, 6.9, 2.8, 1.2])
plt.legend(loc=9, ncol=5, bbox_to_anchor=(0.5, -0.25), borderaxespad=0.,)

_ = [ax1.get_children().pop(x).remove() for x in [8,7,6,4, 3, 2, 1]]
```
